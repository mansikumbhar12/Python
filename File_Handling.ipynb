```python
# Function to write content to a file
def write_to_file(filename, content):
    with open(filename, 'w') as file:
        file.write(content)
    print(f"Content written to {filename}")

# Function to read content from a file
def read_from_file(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
        print("File content:")
        print(content)
        return content
    except FileNotFoundError:
        print(f"{filename} not found!")
        return None

# Function to count the number of words in a file
def count_words_in_file(filename):
    content = read_from_file(filename)
    if content:
        word_count = len(content.split())
        print(f"Total number of words: {word_count}")
        return word_count
    return 0

# Function to search for a word in a file
def search_in_file(filename, word):
    content = read_from_file(filename)
    if content:
        if word in content:
            print(f"'{word}' found in the file.")
        else:
            print(f"'{word}' not found in the file.")
        return word in content
    return False

# Function to find the number of occurrences of a word in a file
def find_word_count(filename, word):
    content = read_from_file(filename)
    if content:
        count = content.count(word)
        print(f"'{word}' occurs {count} time(s) in the file.")
        return count
    return 0
```

```python
# Function to simulate a switch case using a dictionary
def switch_case(choice, filename, word=None, content=None):
    switch = {
        '1': lambda: write_to_file(filename, content),
        '2': lambda: read_from_file(filename),
        '3': lambda: count_words_in_file(filename),
        '4': lambda: search_in_file(filename, word),
        '5': lambda: find_word_count(filename, word)
    }

    # Get the function based on the choice and execute it
    func = switch.get(choice, lambda: print("Invalid choice"))
    return func()

# Function to display options and take user input
def show_options():
    print("\nSelect an option:")
    print("1. Write to file")
    print("2. Read from file")
    print("3. Count words in file")
    print("4. Search for a word in file")
    print("5. Find the number of occurrences of a word in file")
    choice = input("Enter your choice (1-5): ")
    return choice
```

```python
# Main program logic
filename = "example.txt"
content = "This is an example file. It contains multiple lines and words for testing file handling."
word = "example"

while True:
    choice = show_options()

    if choice == '1':
        # Get content from user to write to the file
        content = input("Enter the content to write to the file: ")
        switch_case(choice, filename, content=content)
    elif choice == '2':
        switch_case(choice, filename)
    elif choice == '3':
        switch_case(choice, filename)
    elif choice == '4':
        # Get the word to search for from the user
        word = input("Enter the word to search for: ")
        switch_case(choice, filename, word=word)
    elif choice == '5':
        # Get the word to find its occurrence count
        word = input("Enter the word to count occurrences: ")
        switch_case(choice, filename, word=word)
    else:
        print("Invalid choice. Exiting...")
        break
```
